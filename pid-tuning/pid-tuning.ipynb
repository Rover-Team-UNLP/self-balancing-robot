En el siguiente programa se va a procesar un archivo .csv con la información relevante para entender el funcionamiento de control de nuestro robot. Se utilizara Matplotlib y pandas. En la misma carpeta se encuentran los archivos .csv y ademas un script que permite capturar los datos que la ESP32 imprime por UART y guardarlos en el formato mencionado. 

```python
import pandas as pd
import matplotlib.pyplot as plt

FILENAME = "log_robot_1.csv"

kp = 18
ki = 100
kd = 0.5

try:
    df = pd.read_csv(FILENAME)
except FileNotFoundError:
    print(f'Error al procesar el archivo {FILENAME}')
    
print('Datos Cargados correctamente ')
print(df.columns)

df = df.dropna

df['tiempo_s'] = (df['timestamp'] - df['timestamp'].iloc[0]) / 1000.0

plt.figure(figsize=(12,7))
plt.plot(df['tiempo_s'], df['angulo'], label='Angulo IMU')
plt.plot(df['tiempo_s'], df['setpoint'], label='SetPoint', color='red', linestyle='--')

plt.title(f'Respuesta del Robot Equilibrista (Angulo vs SetPoint) Kp = {kp} | Ki = {ki} | Kd = {kd}')
plt.xlabel('Tiempo [s]')
plt.ylabel('Angulo [°]')
plt.legend()
plt.grid(True)

plt.figure(figsize=(12,7))
plt.plot(df['tiempo_s'], df['salidaPID'], label='Salida PID', color='purple')
plt.title('Esfuerzo de control')
plt.xlabel('Tiempo (s)')
plt.ylabel('Salida PID')
plt.legend()
plt.grid(True)

plt.show()
```
